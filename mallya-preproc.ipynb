```python
def conv(para):   
    accTemp=[]
    for i in df[para]:
        accSplit=i.split("\n ")
        l = []

        for j in accSplit:
            if j=="...":
                continue
            elif j==accSplit[0]:
                j=j[3:-1]
            elif j==accSplit[-1]:
                j=j[2:-2]
            else:
                j=j[2:-1]
            accSplit2=j.split(" ")
            accSplit2 = [float(i) for i in accSplit2 if i != '']
            l.append(accSplit2)

        accTemp.append(l)
    accTemp = np.array(accTemp).reshape(6605, -1)
    print(accTemp.shape)
    return accTemp
```

```python
Accn=conv("Acc")
Gyrn=conv("Gyr")
Magn=conv("Mag")
Barn=conv("Bar")
```

```python
aid=df["ActivityID"]
aid=pd.Series([1 if i>100 else 0 for i in aid ])
aid.value_counts()
```

```python
la = [f"acc{i}_{f}" for i in range(6) for f in ["x", "y", "z"]]
lg = [f"gyr{i}_{f}" for i in range(6) for f in ["x", "y", "z"]]
lm = [f"mag{i}_{f}" for i in range(6) for f in ["x", "y", "z"]]
lb = [f"bar{i}_{f}" for i in range(200) for f in ["1", "2"]]

na = [i for i in range(18)]
nb = [i for i in range(400)]
```

```python
da = {i:j for i, j in zip(na, la)}
dg = {i:j for i, j in zip(na, lg)}
dm = {i:j for i, j in zip(na, lm)}
db = {i:j for i, j in zip(nb, lb)}
```

```python
df_Accn = pd.DataFrame(Accn)
df_Accn.rename(columns=da, inplace=True)
df_Gyrn = pd.DataFrame(Gyrn)
df_Gyrn.rename(columns=dg, inplace=True)
df_Magn = pd.DataFrame(Magn)
df_Magn.rename(columns=dm, inplace=True)
df_Barn = pd.DataFrame(Barn)
df_Barn.rename(columns=db, inplace=True)
df_aid = pd.DataFrame(aid)
df_aid.rename(columns={0 : "fall"}, inplace=True)

df = pd.concat([df_Accn, df_Gyrn, df_Magn, df_Barn, df_aid], axis=1)
```
